```python
import pandas as pd
from datasets import load_dataset
```

```python
dataset = load_dataset("kartashoffv/vedomosti_articles")
```

```python
dataset
```

```python
test_data = pd.concat([dataset["train"].to_pandas(), dataset["test"].to_pandas()])
print(test_data.shape)
print(test_data.head())
```

```python
def write_text_len(row):
    row['len'] = len(row['texts'])
    return row
test_data = test_data.apply(write_text_len, axis=1)
test_data
```

```python
test_data.value_counts('label').plot.bar()
```

```python
sampled_test = test_data.sample(200, random_state=9)
```

```python
sampled_test.value_counts('label').plot.bar()
```

```python
sampled_test.to_csv('test_data.csv', index=False)
```
